```python
import pandas as pd
import numpy as np
```

```python
data_green = pd.read_csv("green_tripdata_2018-01.csv")
data_green = data_green[['VendorID', 'lpep_pickup_datetime', 'lpep_dropoff_datetime', 'passenger_count', 'payment_type']]
data_green = data_green.sample(frac=0.5, replace=False, random_state=1)

print(data_green.shape)

data_green.head(10)
```

```python
data_green1 = pd.read_csv("green_tripdata_2018-02.csv")
data_green1 = data_green1[['VendorID', 'lpep_pickup_datetime', 'lpep_dropoff_datetime', 'passenger_count', 'payment_type']]
data_green1 = data_green1.sample(frac=0.5, replace=False, random_state=1)
print(data_green1.shape)
data_green1.head(10)

```

```python
result = pd.concat([data_green,data_green1], sort=True)
result.shape
```

```python
data_green2 = pd.read_csv("green_tripdata_2018-03.csv")
data_green2 = data_green2[['VendorID', 'lpep_pickup_datetime', 'lpep_dropoff_datetime', 'passenger_count', 'payment_type']]
data_green2 = data_green2.sample(frac=0.5, replace=False, random_state=1)
print(data_green2.shape)
data_green2.head(10)
```

```python
result_green = pd.concat([result,data_green2],sort=True)
result_green.shape
```

```python
result_green.tail(10)
```

```python
result_green.to_csv('result_green.csv',index=False)
```
