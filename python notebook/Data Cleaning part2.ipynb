```python
result = pd.concat([data1,data2])
result.sort_values(['Date','Time'], ascending=True, inplace=True)
```

```python
result['Date'] =  pd.to_datetime(result['Date'], format='%Y-%m-%d')
```

```python
result = (result[(result['Date'] >= '2018-01-01') & (result['Date'] < '2018-04-01')]).reset_index(drop=True)
result.sort_values(['Date','Time'], ascending=True, inplace=True)
result.tail(10)
```
