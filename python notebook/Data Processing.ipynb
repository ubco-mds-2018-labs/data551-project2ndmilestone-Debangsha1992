```python
import numpy as np
import pandas as pd
```

```python
yellow = pd.read_csv('yellow_cab_concise.csv')
green = pd.read_csv('green_cab_concise.csv')
```

```python
yellow.head(10)
```

```python
green.head(10)
```

```python
yellow = yellow[yellow['Trip Month'] < 4]
```

```python
green = green[green['Trip Month'] < 4]
```

```python
new_yellow = pd.DataFrame(yellow.groupby(['Trip Month','payment_type'])['payment_type'].count())
new_yellow
```

```python
new_yellow['Trip Month'] = np.where(new_yellow['Trip Month']==1, 'January', np.where(new_yellow['Trip Month']==2, 'February', 'March'))
```

```python
new_green = pd.DataFrame(green.groupby(['Trip Month','payment_type'])['payment_type'].count())
new_green
```

```python
new_green.to_csv('new_green.csv')
new_yellow.to_csv('new_yellow.csv')
```
